```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score
```

```python

def loglogplot(graph,sort_name,name):
    x = graph["Size"]
    y = graph["Time"]
#     plt.loglog(x,y,'.', basex = 2, basey = 2, markersize = 12)
#     plt.legend()
    plt.xlabel('# of Elements')
    plt.ylabel('time')
    plt.title(sort_name)
    logx, logy = np.log(x), np.log(y)
    m, b = np.polyfit(logx,logy,1)
    fit = np.poly1d((m, b))
    expected_logy = fit(logx)
    r2 = r2_score(logy,expected_logy)
    p = plt.loglog(x,y ,'.', base = 2)
    fit_p = plt.loglog(x[::len(x)-1],(math.e ** expected_logy)[::len(y)-1], '--', base = 2, label = (sort_name + name), markersize = 6, color = p[-1].get_color())
    plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')

    print(sort_name,name,str(': y = {:.4f}*log x + {:.4f}'.format(m,b)))
                       
                       

```

```python
sorting_algorithm = ["merge_sort/","insertion_sort/","shell_sort1/","shell_sort2/","shell_sort3/","shell_sort4/","hybrid_sort1/","hybrid_sort2/","hybrid_sort3/"]
folder = "data/"
perm = ["almost_sorted.csv","reverse.csv","random.csv"]
for i in range(0,9):
    plt.show(block=False)
    for j in range(0,3):

        file_location = folder + sorting_algorithm[i] + perm[j]
        data = pd.read_csv( str(file_location),header=None,names=["Size","Time"], index_col=False)
        graph = data.dropna()
        loglogplot(graph,sorting_algorithm[i],perm[j])
    

    
```

```python

for j in range(0,3):
    data = pd.read_csv( str("data/merge_sort/"+perm[j]),header=None,names=["Size","Time"], index_col=False)
    graph = data.dropna()
    loglogplot(graph,"merge-",perm[j])
for j in range(0,3):
    data = pd.read_csv( str("data/insertion_sort/"+perm[j]),header=None,names=["Size","Time"], index_col=False)
    graph2 = data.dropna()
    loglogplot(graph2,"insertion-",perm[j])
plt.title("insertion & merge")
```
