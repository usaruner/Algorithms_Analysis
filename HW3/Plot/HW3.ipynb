```python
def linlogplot(graph,sort_name,name):
    x = graph["Size"]
    y = graph["Waste"]
#     plt.loglog(x,y,'.', basex = 2, basey = 2, markersize = 12)
#     plt.legend()
    plt.xlabel('# of Elements')
    plt.ylabel('num')
    plt.title(sort_name)
    plt.plot(x,y,'.')
    
```

```python
sorting_algorithm = ["diameter_","cluster_"]
folder = "final_data/"
perm = "random.csv"
for i in range(0,len(sorting_algorithm)):
    plt.show(block=False)
    file_location = folder + sorting_algorithm[i] + perm
    data = pd.read_csv( str(file_location),header=None,names=["Size","Waste"], index_col=False)
    graph = data.dropna()
    linlogplot(graph,sorting_algorithm[i],perm)
```
