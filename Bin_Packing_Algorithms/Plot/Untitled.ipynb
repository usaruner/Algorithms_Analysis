```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score
```

```python

def loglogplot(graph,sort_name,name):
    x = graph["Size"]
    y = graph["Waste"]
#     plt.loglog(x,y,'.', basex = 2, basey = 2, markersize = 12)
#     plt.legend()
    plt.xlabel('# of Elements')
    plt.ylabel('waste')
    plt.title(sort_name)
    logx, logy = np.log(x), np.log(y)
    m, b = np.polyfit(logx,logy,1)
    fit = np.poly1d((m, b))
    expected_logy = fit(logx)
    r2 = r2_score(logy,expected_logy)
    p = plt.loglog(x,y ,'.', base = 2)
    fit_p = plt.loglog(x[::len(x)-1],(math.e ** expected_logy)[::len(y)-1], '--', base = 2, label = (sort_name + name), markersize = 6, color = p[-1].get_color())
    plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')

    print(sort_name,name,str(': y = {:.4f}*log x + {:.4f}'.format(m,b)))
                       
                       

```

```python
sorting_algorithm = ["best_fit_","best_fit_decreasing_","first_fit_decreasing_","first_fit_","next_fit_"]
folder = "data/"
perm = "random.csv"
for i in range(0,len(sorting_algorithm)):
    plt.show(block=False)
    file_location = folder + sorting_algorithm[i] + perm
    data = pd.read_csv( str(file_location),header=None,names=["Size","Waste"], index_col=False)
    graph = data.dropna()
    loglogplot(graph,sorting_algorithm[i],perm)
```

```python
data = pd.read_csv( str("data/best_fit_random.csv"),header=None,names=["Size","Waste"], index_col=False)
graph = data.dropna()
loglogplot(graph,"BF-",perm)

data = pd.read_csv( str("data/first_fit_random.csv"),header=None,names=["Size","Waste"], index_col=False)
graph2 = data.dropna()
loglogplot(graph2,"FF-",perm)

data = pd.read_csv( str("data/next_fit_random.csv"),header=None,names=["Size","Waste"], index_col=False)
graph3 = data.dropna()
loglogplot(graph3,"NF-",perm)


plt.title("NF vs FF vs BF")
```
